```python
# Q3 Statistical Analysis 
# Compute the 25th, 50th, and 75th percentiles of the weight column using NumPy.

# Extract the "weight" column as a NumPy array
weights = np.array(df["weight"])

# Calculate the 25th, 50th, and 75th percentiles
percentile_25 = np.percentile(weights, 25)
percentile_50 = np.percentile(weights, 50)  # This is also the median
percentile_75 = np.percentile(weights, 75)

# Print the results
print(f"25th Percentile: {percentile_25}")
print(f"50th Percentile (Median): {percentile_50}")
print(f"75th Percentile: {percentile_75}")
```

```python
#Q4 Convert the acceleration column into a NumPy array and normalize its values (scale between 0 and 1).

# Convert the "acceleration" column to a NumPy array
acceleration = np.array(df["acceleration"])

# Normalize the values between 0 and 1 using min-max scaling
acceleration_min = np.min(acceleration)
acceleration_max = np.max(acceleration)

normalized_acceleration = (acceleration - acceleration_min) / (acceleration_max - acceleration_min)

# Print the normalized values
print("Normalized Acceleration Values:")
print(normalized_acceleration)
```

```python
#Q5 Convert the "horsepower" column to a NumPy array
horsepower = np.array(df["horsepower"])

# Handle missing data by replacing NaN values with the mean of the column
horsepower_mean = np.nanmean(horsepower)  # np.nanmean ignores NaN values while calculating the mean
horsepower[np.isnan(horsepower)] = horsepower_mean  # Replace NaNs with the mean

# Increase all horsepower values by 10%
updated_horsepower = horsepower * 1.10  # 10% increase

# Print the updated horsepower values
print("Updated Horsepower Values (increased by 10%):")
print(updated_horsepower)
```

```python
#Q7 Matrix Operations 
# Create a 2D NumPy array containing the columns mpg, horsepower, and weight. 
# Compute the dot product of this matrix with a given vector [1, 0.5, -0.2]. 
import numpy as np
import pandas as pd

# Load the dataset
df = pd.read_csv("C:/Users/anupa/OneDrive/Desktop/ICTA/auto-mpg dsa.csv")

# Extract the relevant columns and create a 2D NumPy array
data_matrix = np.array(df[["mpg", "horsepower", "weight"]])

# Define the given vector
vector = np.array([1, 0.5, -0.2])

# Compute the dot product of the data_matrix with the given vector
dot_product = np.dot(data_matrix, vector)

# Print the result
print("Dot product of the matrix with the vector:")
print(dot_product)
```

```python
#Q8 Use NumPy to sort the cars by model_year in descending order and display the first five car names.

# Convert "model_year" and "car name" columns to NumPy arrays
model_year = np.array(df["modelyear"])
car_names = np.array(df["car name"])

# Get the indices that would sort the model_year in descending order
sorted_indices = np.argsort(model_year)[::-1]  # Reversing the order for descending

# Sort the car names using the sorted indices
sorted_car_names = car_names[sorted_indices]

# Display the first five car names
print("First five cars sorted by model_year in descending order:")
print(sorted_car_names[:5])
```

```python
# Q9 Correlation 
# Compute the Pearson correlation coefficient between mpg and weight using NumPy.?

# Drop rows with missing values in the relevant columns (if any)
df = df[['mpg', 'weight']].dropna()

# Compute the Pearson correlation coefficient
correlation_matrix = np.corrcoef(df['mpg'], df['weight'])
pearson_corr = correlation_matrix[0, 1]

print(f"Pearson correlation coefficient between mpg and weight: {pearson_corr:.4f}")
```

```python
#Q10 Calculate the mean mpg for cars grouped by the number of cylinders using NumPy techniques. 


import numpy as np
import pandas as pd

# Load the dataset
df = pd.read_csv("C:/Users/anupa/OneDrive/Desktop/ICTA/auto-mpg dsa.csv")

# Drop rows with missing values in the relevant columns
df = df[['mpg', 'cylinders']].dropna()

# Convert to NumPy arrays
mpg = df['mpg'].to_numpy()
cylinders = df['cylinders'].to_numpy()

# Find unique cylinder counts
unique_cylinders = np.unique(cylinders)

# Compute mean mpg for each cylinder group
for cyl in unique_cylinders:
    mean_mpg = np.mean(mpg[cylinders == cyl])
    print(f"Mean mpg for {int(cyl)} cylinders: {mean_mpg:.2f}")
```
